## Let's first look without adding more columns to our dataset.

```python
from sklearn import datasets
import pandas as pd
boston = datasets.load_boston()
```

```python
df = pd.DataFrame(boston.data)
```

```python
df.columns
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(df, boston.target, test_size=0.2, random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape,y_test.shape
```

```python
alg_boston = LinearRegression()
alg_boston.fit(x_train, y_train)
```

```python
y_pred = alg_boston.predict(x_test)
```

```python
alg_boston.score(x_test, y_test)
```

## Here, the score calculated above is 0.7634....... Let's now try to add more columns to our data set and see if this score improves or not.

# Adding more columns to our dataset is also known as finding complex boundaries for our given dataset.

```python
from sklearn import datasets
import pandas as pd
boston1 = datasets.load_boston()
```

```python
df1 = pd.DataFrame(boston.data)
```

```python
df1.columns
```

```python
i_j = 13
for i in range(13):
    for j in range(i+1):
        df1[i_j] = df1[i] * df1[j]
        i_j += 1
    
```

#### Here we have added more columns i.e multiplication of each column with all other columns.

```python
df1
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(df1, boston1.target, test_size=0.2, random_state=1)
```

```python
x_train1.shape, x_test1.shape, y_train1.shape, y_test1.shape
```

```python
alg_boston1 = LinearRegression()
alg_boston1.fit(x_train1, y_train1)
```

```python
y_pred1 = alg_boston1.predict(x_test1)
```

```python
alg_boston1.score(x_test1, y_test1)
```

# Here, the score is far better than the score calculated before.
